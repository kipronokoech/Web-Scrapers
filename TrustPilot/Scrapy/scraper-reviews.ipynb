## Web scrap all pages of the reviews

There are so many pages inthe site that we need to scrap. Each page contains a unique url but with clear patterns so that we are able to loop through each page. Pages will have urls that apper as:

- https://www.trustpilot.com/review/www.worldremit.com
- https://www.trustpilot.com/review/www.worldremit.com?page=2
- https://www.trustpilot.com/review/www.worldremit.com?page=3
- https://www.trustpilot.com/review/www.worldremit.com?page=4 etc

The difference between the link to one page to another is the page number therefore we will be able to loop through the pages while scrapping the information we need.

```python
f = open("./data/reviews.json","r")
def scrap_page(url_link):
    #opening up connection, scrapping the page
    client = ureq(url_link)
    raw_html = client.read()
    client.close()
    #html parsing
    page_soup = bs(raw_html,"html.parser")
    #Glabs each review card, just within  the page
    containers = page_soup.findAll("div",{"class":"review-card"})
    #we can now loop over all containers
    for index,container in enumerate(containers):
        script = container.find('script')
        content = script.contents[0]
        f.write(content+"\n")


# TO-DO : check on how to count the number of available pages without defining it explicitly.

for i in range(1,1824): #by the time of making this notebook there were 1824 pages 
    my_url = "https://www.trustpilot.com/review/www.worldremit.com?page={}".format(i)
    scrap_page(my_url)
    print(my_url," scraped.")

f.close()
```
